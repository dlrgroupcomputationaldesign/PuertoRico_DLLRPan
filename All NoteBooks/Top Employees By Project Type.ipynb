```python
planifi['4 Week Hours Forecast']=planifi[['03/22/2021', '03/29/2021', '04/05/2021', '04/12/2021']].sum(axis=1)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))

sns.set_theme(style="whitegrid")

tips = sns.load_dataset("tips")

ax = sns.boxplot(data=planifi[planifi["Planned Next 4"]!=160],
                 orient="h",
                 x="4 Week Hours Forecast",
                 y='DISCIPLINE')

ax.axvline(160,lw=2,color='firebrick',ls='--')

```

```python
data['EMPLOYEE']=data['First Name']+" "+data['Last Name']
```

```python
data[['EMPLOYEE','Discipline','Sector Name','Charged Hours']].\
    sort_values(['Charged Hours'],ascending=False).\
    sort_values(['Discipline','Sector Name']).\
    groupby(['Discipline','Sector Name']).\
    head(10).\
    set_index(['EMPLOYEE']).\
    join(planifi.set_index(['EMPLOYEE'])['4 Week Hours Forecast']).\
    reset_index()
```

```python
data[['EMPLOYEE','Discipline','Sector Name','Charged Hours']].\
    sort_values(['Charged Hours'],ascending=False).\
    sort_values(['Discipline','Sector Name']).\
    groupby(['Discipline','Sector Name']).\
    head(10).\
    set_index(['EMPLOYEE']).\
    join(planifi.set_index(['EMPLOYEE'])['4 Week Hours Forecast']).\
    reset_index().to_csv('C:\\Users\dmckenzie\OneDrive - DLR Group\TopByDiscSector.csv')
```

```python
data[['EMPLOYEE','Discipline','Sector Name','Charged Hours']].\
    sort_values(['Charged Hours'],ascending=False).\
    sort_values(['Discipline','Sector Name']).\
    groupby(['Discipline','Sector Name']).\
    tail(10).\
    set_index(['EMPLOYEE']).\
    join(planifi.set_index(['EMPLOYEE'])['4 Week Hours Forecast']).\
    reset_index()
```

```python
data[['EMPLOYEE','Discipline','Sector Name','Charged Hours']].\
    sort_values(['Charged Hours'],ascending=False).\
    sort_values(['Discipline','Sector Name']).\
    groupby(['Discipline','Sector Name']).\
    tail(10).\
    set_index(['EMPLOYEE']).\
    join(planifi.set_index(['EMPLOYEE'])['4 Week Hours Forecast']).\
    reset_index().to_csv('C:\\Users\dmckenzie\OneDrive - DLR Group\BottomByDiscSector.csv')
```

```python
data2=data[['EMPLOYEE','Discipline','Sector Name','Charged Hours']].\
    set_index(['EMPLOYEE']).\
    join(planifi.set_index(['EMPLOYEE'])['4 Week Hours Forecast']).\
    reset_index()
data2['StarVar']=(data2['Charged Hours']+1)/(data2['4 Week Hours Forecast']+1)
data2.sort_values(['StarVar'],ascending=False).\
    sort_values(['Discipline','Sector Name']).\
    groupby(['Discipline','Sector Name']).\
    head(10)[0:50]
```
