```python
Taxes=pd.pivot_table(DFtax,
               values='Amount',
               index=['State','Year','Numeric Month'],
               columns=['Tax Type'],
               aggfunc=np.sum).\
    reset_index()
```

```python
Meltprice=pd.melt(DFprice, 
                  id_vars=["RegionID",'RegionName','RegionType','StateName']).\
    rename(columns={'variable':'Date','value':'MedianPrice'})
```

```python
Meltlist=pd.melt(DFlist,
                 id_vars=["RegionID",'RegionName','StateName']).\
    rename(columns={'variable':'Date','value':'Listings'})
```

```python
DFpl=Meltprice.set_index(['RegionID','Date']).\
    join(Meltlist.drop(columns=['RegionName','StateName']).\
             set_index(['RegionID','Date'])).\
    reset_index()
```

```python
from plotnine import *
```

```python
DFpl[['MedianPrice','Listings']].corr()
```

```python
DFpl.groupby('StateName')[['MedianPrice','Listings']].corr().query('MedianPrice!=1').drop(columns='Listings')
```

```python
DFpl.groupby(DFpl.Date.dt.month)[['MedianPrice','Listings']].corr().query('MedianPrice!=1').drop(columns='Listings')
```

```python
DFpl.groupby(DFpl.Date.dt.month)[['MedianPrice','Listings']].corr().\
    query('MedianPrice!=1').\
    reset_index().\
    drop(columns=['Listings','level_1']).\
    set_index('Date').\
    plot()
```

```python
DFpl.groupby([DFpl.Date.dt.month,'StateName'])[['MedianPrice','Listings']].corr().\
    query('MedianPrice!=1').\
    reset_index().\
    drop(columns=['Listings','level_2']).groupby(['StateName']).plot(x='Date',y='MedianPrice')
```

```python
CorrFluc=DFpl.groupby([DFpl.Date.dt.month,'StateName'])[['MedianPrice','Listings']].corr().\
    query('MedianPrice!=1').\
    reset_index().\
    drop(columns=['Listings','level_2'])
```

```python
(
ggplot(CorrFluc,aes(x='Date',y='MedianPrice',color='StateName'))+
    geom_line()+
    geom_point()+
    theme_bw()+
    facet_wrap("StateName")
)
```
