```python
NumDat.apply(pd.to_numeric, errors='ignore').describe().T
```

```python
NumDat=NumDat.apply(pd.to_numeric, errors='ignore')
```

```python
for i in NumDat.columns:
    if max(NumDat[i])==1:
        NumDat[i]=NumDat[i]+(np.random.normal(0,.05, size=len(NumDat)))
    
```

```python
import umap
```

```python
reducer = umap.UMAP( n_components=3,random_state=0)
embedding = reducer.fit_transform(NumDat)
```

```python
bedded2=pd.DataFrame({'X1':embedding[:,0],
                      "X2":embedding[:,1],
                      "X3":embedding[:,2]})
import hdbscan
clusterer = hdbscan.HDBSCAN(min_cluster_size=40)
cluster_labels = clusterer.fit_predict(bedded2[['X1','X2','X3']])
#######################################################
bedded2['cluster']=cluster_labels.astype('str')
#######################################################
```

```python
NumDat2=pd.concat([bedded2[['X1','X2','X3']].reset_index(drop=True),NumDat.reset_index(drop=True)],axis=1)
```

```python
C1=NumDat2.drop(columns=(['X2','X3'])).corr().abs()[['X1']].reset_index().sort_values('X1',ascending=False).rename(columns={"X1":"Cor"})
```

```python
C2=NumDat2.drop(columns=(['X1','X3'])).corr().abs()['X2'].reset_index().sort_values('X2',ascending=False).rename(columns={"X2":"Cor"})
```

```python
C3=NumDat2.drop(columns=(['X1','X2'])).corr().abs()['X3'].reset_index().sort_values('X3',ascending=False).rename(columns={"X3":"Cor"})
```

```python
C1.append(C2).append(C3).groupby('index')['Cor'].max().reset_index().sort_values('Cor',ascending=False).query('Cor>.00 & Cor<1').reset_index(drop=True)
```

```python
NewNums=list(C1.append(C2).append(C3).groupby('index')['Cor'].max().reset_index().sort_values('Cor',ascending=False).query('Cor>.07 & Cor<1').reset_index(drop=True)['index'])
```

```python
NumDat3=NumDat[NewNums]
reducer = umap.UMAP( n_components=3,random_state=0)
embedding = reducer.fit_transform(NumDat3)
```

```python
bedded2=pd.DataFrame({'X1':embedding[:,0],
                      "X2":embedding[:,1],
                      "X3":embedding[:,2]})
import hdbscan
clusterer = hdbscan.HDBSCAN(min_cluster_size=25)
cluster_labels = clusterer.fit_predict(bedded2[['X1','X2','X3']])
#######################################################
bedded2['cluster']=cluster_labels.astype('str')
```
