```python
Narratives=pd.read_csv('C:/Users\dmckenzie\OneDrive - DLR Group\Desktop\Project Narritives Jan 2022.csv')
```

```python
Narratives['Narrative']=Narratives['Narrative'].str.lower()
```

```python
import nltk 
```

```python
def wordmatcher(tokenlist,keys):
    tokens = [x for x in keys if x in str(tokenlist)]
    return (tokens)
```

```python
from nltk.corpus import stopwords
stops=list(stopwords.words('english'))+['um','know','try','trying','get',
                                        "that's","think",'say',
                                        "there's",'lot','go','back',
                                       "would",'yeah',"i'm",'also','even','way','got',
                                       'things','well','like','kind','something','actually','still','5',"one",
                                       'might','maybe','could','okay',"we've","they're","we're",' ']
def tokennamefilter(tokenlist,minlen=2):
    tokens = [x for x in tokenlist if x not in stops]
    return (tokens)
```

```python
values['K2']=values['Key Words'].str.lower().str.split(',').apply(tokennamefilter).apply(','.join).str.split(', ')
```

```python
values
```

```python
for i in values['Concept name']:
    vallist=(values[values['Concept name']==i]['K2'].reset_index(drop=True)[0])
    a=[]
    for j in Narratives['Narrative']:
        b=wordmatcher(j,vallist)
        a.append(b)
    Narratives[i]=a
    Narratives[i]=Narratives[i].apply(' '.join)
```
