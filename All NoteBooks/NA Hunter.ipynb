```python
import pyodbc 
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
```

```python
#This report serves two goals.

#1. Find what custome fields could be axed with minimal pain on the part of the user 
#2. Find what fields users like based on how often they will them in

#First, we look at NA distributions by table and peak at columns users aren't filling in

#Second, we look at columns that filled in "pretty often", omitting columns that are always filled in,
#as those columns are liekly keys or required columns
```

```python
matplotlib.style.use('seaborn')
```

```python
conn = pyodbc.connect('Driver={SQL Server};'
                      'Server=VisionDB;'
                      'Database=Sandbox_Vision;'
                      'Trusted_Connection=yes;')
cursor = conn.cursor()
```

```python
query = """Select *  FROM [Sandbox_Vision].[dbo].[ClientCustomTabFields]"""
RC = pd.read_sql_query(query,conn)
clientsna=round(RC.isna().sum()/len(RC),3).sort_values(ascending=False)
clientsna=clientsna.reset_index().rename(columns={0:'Percent NA','index':'Column_Clients'})
clientsna.plot(title='Column NA Distribution').title.set_size(20)
```

```python
clientsna[clientsna['Percent NA']>.7]
```
